### **TASK-1: Prediction Using Supervised ML**
In this regression task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variable

### Author- Sarven Lad.


```python
# importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
# Reading the Data 
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
data.head(10)
```

```python
# Check if there any null value in the Dataset
data.isnull == True
```

##### **There is no null value in the given Dataset**

```python
sns.set_style('darkgrid')
sns.scatterplot(y= data['Scores'], x= data['Hours'])
plt.title('Marks Vs Study Hours',size=20)
plt.ylabel('Marks Percentage', size=15)
plt.xlabel('Hours Studied', size=15)
plt.show()
```

##### **From the above scatter plot there looks to be correlation between the 'Mark  Percentage' and 'Hours Studied', Lets plot a regression line to confirm the correlation**.

```python
sns.regplot(x= data['Hours'], y= data['Scores'])
plt.title('Regression Plot',size=20)
plt.ylabel('Marks Percentage', size=15)
plt.xlabel('Hours Studied', size=15)
plt.show()
print(data.corr())
```

##### **It is confirmed that the variables are positively correlated.**

## Training the Model

```python
# Defining X and y from the Data
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values

# Spliting the Data in two
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
regression = LinearRegression()
regression.fit(train_X, train_y)
print("---------Model Trained---------")
```

### Predicting the percentage of marks

```python
pred_y = regression.predict(val_X)
prediction = pd.DataFrame({'Hours': [i[0] for i in val_X], 'Predicted Marks': [k for k in pred_y]})
prediction
```

### Comparing the Predicted Marks with the Actual Marks

```python
compare_scores = pd.DataFrame({'Actual Marks': val_y, 'Predicted Marks': pred_y})
compare_scores
```

### Visually Comparing the Predicted Marks with the Actual Marks

```python
plt.scatter(x=val_X, y=val_y, color='blue')
plt.plot(val_X, pred_y, color='Black')
plt.title('Actual vs Predicted', size=20)
plt.ylabel('Marks Percentage', size=15)
plt.xlabel('Hours Studied', size=15)
plt.show()
```

```python
# Calculating the accuracy of the model
print('Mean absolute error: ',mean_absolute_error(val_y,pred_y))
```

##### **A low value of Mean Absolute Error indicates a minimal likelihood of errors or inaccuracies in the model's predictions, suggesting a high level of accuracy in forecasting**

## What will be the predicted score of a student if he/she studies for 9.25 hrs/ day?

```python
hours = [9.25]
answer = regression.predict([hours])
print("Score = {}".format(round(answer[0],3)))
```

##### **Based on the regression model, if a student dedicates 9.25 hours per day to studying, it is predicted that he or she will achieve a score of approximately 93.89 marks**
